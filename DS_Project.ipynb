```python
# Cleaning of Dataset : In this stage start cleaning our dataset in order to continue the analysis.
# Handaling Missing Values
```

```python
# Drop the serial no. as it is no use . Because index no. and serial no. is same .
data.drop('s.no', axis=1, inplace=True)
data
```

```python
#Checking the missing values
data.isnull()
```

```python
# Total no. of missing Value
data.isnull().sum()
```

```python
# sorting the missing value in decsending order
total=data.isnull().sum().sort_values(ascending=False)
total
```

```python
# Percentage of missing values per column
missing_percent=data.isnull().sum().sort_values(ascending=False)/len(data)
missing_percent
```

```python
missing_data=pd.concat([total,missing_percent],axis=1,keys=['Total','Percentage'])
missing_data
```

```python
# Dealing with missing data
data=data.dropna()
data
```

```python
# Let's calculate the percentage of class_name of each columns
data.class_name.value_counts(normalize=True)
# plot the bar graph of % of class name category
data.class_name.value_counts(normalize=True).plot.barh()
```

```python
# Countplot is used to show the total count of observation in the department_name
plt.figure(figsize=(10,7))
sns.countplot(x='department_name',data=data,order=data['department_name'].value_counts().index)
plt.xticks(rotation=60)
plt.show()
```

```python
# Histogram is used to count the no.of obseration in Rating colunm that fall in each bin .
sns.histplot(x='rating',data=data , kde=True)
```

```python
# Pie plot for the department name
plt.figure(figsize=(10,8))
data['department_name'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# Barplot for division_name  of cloth .
plt.figure(figsize=(8,4))
data['division_name'].value_counts().plot.bar(color='red')
plt.show()
```

```python
#Scatter plot
# Plot the sactter plot of the Age and  in data
plt.scatter(data.age,data.alike_feedback_count,color='green',marker='^')
plt.xlabel('Age')
plt.ylabel('Feedback Count')
plt.show()
```

```python
plt.scatter(x='department_name',y='age',data=data,c ="pink",
            linewidths = 4,marker ="*")
plt.xlabel('Department name of Cloth')
plt.ylabel('Age of person')
plt.show()
```

```python
# Barplot Between Division name and Department name
sns.barplot(x='division_name',y='alike_feedback_count',data=data,palette='rainbow',alpha=1)
plt.title('Relationship b/w Division of cloth and Feedback count')
plt.show()
```

```python
# Using groupby function
data.groupby('division_name')[['rating','alike_feedback_count']].sum().plot.bar(color=['green','skyblue'],
                                                                                alpha=0.7,figsize=(6,4))
plt.xlabel('Division name of cloth')
plt.ylabel('Rating and Feedback')
plt.show()
```
