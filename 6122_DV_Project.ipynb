```python
#By using seaborn library plotting countplot
#countplot() is used to show the counts of observations in each categorical bin using bars
#count plot on single categorical variable
#The ginraph represent how many Male and Female active customer are available in the different city.

import seaborn as sns
%matplotlib notebook
import seaborn as sns
sns.countplot(x='Gender', hue='City' , data=df, palette='viridis_r')
```

```python
#Histogram are visualization tools that represent the distribution of a set of continuous data.
#In a histogram the data is divided into a set of bins(usually on the x-axis).
#The graph represent the 5% Tax of product That are gained by company with respect to Customer.
#Parameters are x axis data which is Tax 5% , bins is 6, and edge color of graph is black.

import seaborn as sns
%matplotlib notebook
sns.histplot(x='Tax 5%',bins=6,edgecolor='k',data=df)
```

```python
#Box plot is the visual representation of the depicting groups of numerical data through their quartiles.
#Boxplot of Branch with respect Quantity
# Bottom black horizontal line of box plot is minimum value
# First black horizontal line is 1st quartile 
# Second black horizontal line is 2nd quartile or median 
# Thrid line is 3rd quartile or
# Top black line is maximum value of data
# Small triangular shape of green box is outlier

import seaborn as sns
%matplotlib notebook
sns.set_style('whitegrid')
sns.boxplot(x='Branch',y='Quantity',data=df, showmeans=True, saturation=0.95 ,palette='RdPu_r',width=0.5)
```

```python
# Barplot is basically used to aggregate the categorical data according to some methods and by default it's mean.
# To use this plot we choose a categorical column for the x-axis and 
# a numerical column for the y-axis and we see that it creates a plot taking a mean per categorical value.
# City vs Total barplot
# here City are in Categorical column on x- axis
# and Total is a Numerical column on y-axis

import seaborn as sns
%matplotlib notebook
sns.barplot(x='City',y='Total',data=df)
```

```python
# Strip Plot is drawn on its own . It is good complement to a boxplot.
# It is used to draw a scatter plot based on the category.
# Gender vs Total Stripplot
# Here's representing how many male and female spend total Amount on the product using Stripplot

import seaborn as sns
%matplotlib notebook
sns.stripplot(x='Gender',y='Total',data=df,jitter=0.2,hue='Product line',palette='Dark2')
```

```python
import matplotlib.pyplot as plt
sns.jointplot(x='Quantity',y='gross income',data=df,kind='reg')
plt.show()
```
