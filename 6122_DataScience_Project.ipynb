```python
# Using groupby function
data.groupby('division_name')[['rating','alike_feedback_count']].sum().plot.bar(color=['green','skyblue'],
                                                                                alpha=0.7,figsize=(6,4))
plt.xlabel('Division name of cloth')
plt.ylabel('Rating and Feedback')
plt.show()
```

```python
# groupby department_name to find the mean of rating separetely
data.groupby('department_name')['rating'].mean()
```

```python
# groupby department_name to find the median of rating for each separetely
data.groupby('department_name')['rating'].median()
```

```python
# Boxplot between department and Rating for each 
sns.boxplot(x='department_name',y='rating',data=data)
plt.show()
```

```python
# groupby the class_name to find the sum/total of rating and alike_feedback_count separetely .
df=data.groupby('class_name')[['rating','alike_feedback_count']].sum().sort_values(by='rating',ascending=False)
df[:].plot.bar(color=['orange','brown'],alpha=0.8,figsize=(8,6))
plt.title('Rating and feedback of Each Cloth')
plt.xlabel('Cloth name')
plt.ylabel('Rating and Feedback')
plt.show()
```
